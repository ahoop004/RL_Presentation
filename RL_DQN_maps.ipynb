```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

import torch
import torch.nn as nn
import torch.optim as optim
from collections import deque
import random

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
def load_mazes(filepath):
    with open(filepath) as f:
        content = f.read()

    maps = []
    for block in content.split('-' * 60):
        block = block.strip()
        if not block:
            continue
        ns = {}
        try:
            exec(block, {'np': np}, ns)
            maps.append({
                'maze':  ns['maze'],
                'start': ns['start'],
                'goal':  ns['goal'],
            })
        except Exception:
            continue
    return maps


maps = load_mazes('mazes.txt')
print(f"Loaded {len(maps)} maze(s)")
for i, m in enumerate(maps):
    s = m['maze'].shape
    print(f"  [{i}] {s[0]}x{s[1]}  start={m['start']}  goal={m['goal']}")
```

```python
num_episodes = 50000
max_steps_per_episode = 1000
gamma = 0.999
epsilon = 1.0
epsilon_decay = 0.99995
epsilon_min = 0.01

batch_size = 512
replay_buffer_size = 50000
target_update_freq = 10
lr = 0.001

reward_distance_scale = 0.0
reward_fire = -10
reward_goal = 50
reward_step = -1

actions = [(0, -1), (0, 1), (-1, 0), (1, 0)]
```

```python
class DQN(nn.Module):
    def __init__(self, input_size=8, hidden_size=128, output_size=4):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_size, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, output_size)
        )

    def forward(self, x):
        return self.net(x)


policy_net = DQN().to(device)
target_net = DQN().to(device)
target_net.load_state_dict(policy_net.state_dict())
target_net.eval()

optimizer = optim.Adam(policy_net.parameters(), lr=lr)

replay_buffer = deque(maxlen=replay_buffer_size)

# Set by the training loop at the start of each episode
current_maze  = None
current_start = None
current_goal  = None


def state_to_tensor(state):
    r, c = state
    max_idx = current_maze.shape[0] - 1
    def passable(nr, nc):
        return float(0 <= nr < current_maze.shape[0] and 0 <= nc < current_maze.shape[1] and current_maze[nr, nc] == 0)
    return [
        r / max_idx,
        c / max_idx,
        passable(r, c - 1),        # left
        passable(r, c + 1),        # right
        passable(r - 1, c),        # up
        passable(r + 1, c),        # down
        current_goal[0] / max_idx, # goal row
        current_goal[1] / max_idx, # goal col
    ]


def sample_batch():
    batch = random.sample(replay_buffer, batch_size)
    states, acts, rews, next_states, dones = zip(*batch)
    return (
        torch.tensor(states, dtype=torch.float32).to(device),
        torch.tensor(acts, dtype=torch.long).to(device),
        torch.tensor(rews, dtype=torch.float32).to(device),
        torch.tensor(next_states, dtype=torch.float32).to(device),
        torch.tensor(dones, dtype=torch.float32).to(device)
    )
```

```python
def is_valid(pos):
    r, c = pos
    if r < 0 or r >= current_maze.shape[0]:
        return False
    if c < 0 or c >= current_maze.shape[1]:
        return False
    if current_maze[r, c] == 1:
        return False
    return True


def choose_action(state):
    if np.random.random() < epsilon:
        return np.random.randint(len(actions))
    state_t = torch.tensor([state_to_tensor(state)], dtype=torch.float32).to(device)
    with torch.no_grad():
        return policy_net(state_t).argmax(dim=1).item()
```

```python
def manhattan(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])


rewards_all_episodes = []

for episode in range(num_episodes):
    # Pick a random maze for this episode
    global current_maze, current_start, current_goal
    m = random.choice(maps)
    current_maze  = m['maze']
    current_start = m['start']
    current_goal  = m['goal']

    state = current_start
    total_rewards = 0
    done = False

    for _ in range(max_steps_per_episode):
        if done:
            break

        action_index = choose_action(state)
        action = actions[action_index]
        next_state = (state[0] + action[0], state[1] + action[1])

        if not is_valid(next_state):
            reward = reward_fire
            next_state = state
            done = True
        elif next_state == current_goal:
            reward = reward_goal
            done = True
        else:
            dist_before = manhattan(state, current_goal)
            dist_after  = manhattan(next_state, current_goal)
            reward = reward_step + (dist_before - dist_after) * reward_distance_scale

        replay_buffer.append((
            state_to_tensor(state),
            action_index,
            reward,
            state_to_tensor(next_state),
            float(done)
        ))

        if len(replay_buffer) >= batch_size:
            states_t, actions_t, rewards_t, next_states_t, dones_t = sample_batch()

            q_values = policy_net(states_t).gather(1, actions_t.unsqueeze(1)).squeeze(1)

            with torch.no_grad():
                next_q = target_net(next_states_t).max(dim=1).values
                targets = rewards_t + gamma * next_q * (1 - dones_t)

            loss = nn.functional.smooth_l1_loss(q_values, targets)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        state = next_state
        total_rewards += reward

    epsilon = max(epsilon_min, epsilon * epsilon_decay)

    if episode % target_update_freq == 0:
        target_net.load_state_dict(policy_net.state_dict())

    rewards_all_episodes.append(total_rewards)

    if (episode + 1) % 100 == 0:
        avg_reward = np.mean(rewards_all_episodes[-100:])
        print(f"Episode {episode + 1:>5}/{num_episodes} | Avg Reward (last 100): {avg_reward:>8.1f} | Epsilon: {epsilon:.3f}")
```

```python
def get_optimal_path(policy_net, start, goal, actions, maze, max_steps=1000):
    path = [start]
    state = start
    visited = set()

    policy_net.eval()
    with torch.no_grad():
        for _ in range(max_steps):
            if state == goal:
                break
            visited.add(state)

            state_t = torch.tensor([state_to_tensor(state)], dtype=torch.float32).to(device)
            q_vals = policy_net(state_t).squeeze(0).cpu().numpy()

            best_action = None
            best_value = -float('inf')
            for idx, move in enumerate(actions):
                next_state = (state[0] + move[0], state[1] + move[1])
                if (0 <= next_state[0] < maze.shape[0] and
                    0 <= next_state[1] < maze.shape[1] and
                    maze[next_state] == 0 and
                        next_state not in visited):
                    if q_vals[idx] > best_value:
                        best_value = q_vals[idx]
                        best_action = idx

            if best_action is None:
                break

            move = actions[best_action]
            state = (state[0] + move[0], state[1] + move[1])
            path.append(state)

    return path


m = maps[0]
current_maze  = m['maze']
current_start = m['start']
current_goal  = m['goal']

optimal_path = get_optimal_path(policy_net, current_start, current_goal, actions, current_maze)
```

```python
def plot_maze_with_path(path):
    cmap = ListedColormap(['#eef8ea', '#a8c79c'])

    plt.figure(figsize=(8, 8))
    plt.imshow(current_maze, cmap=cmap)

    plt.scatter(current_start[1], current_start[0], marker='o', color='#81c784', edgecolors='black',
                s=200, label='Start (Robot)', zorder=5)
    plt.scatter(current_goal[1], current_goal[0], marker='*', color='#388e3c', edgecolors='black',
                s=300, label='Goal (Diamond)', zorder=5)

    rows, cols = zip(*path)
    plt.plot(cols, rows, color='#60b37a', linewidth=4,
             label='Learned Path', zorder=4)

    plt.title('Reinforcement Learning: Robot Maze Navigation')
    plt.gca().invert_yaxis()
    plt.xticks(range(current_maze.shape[1]))
    plt.yticks(range(current_maze.shape[0]))
    plt.grid(True, alpha=0.2)
    plt.legend()
    plt.tight_layout()
    plt.show()


plot_maze_with_path(optimal_path)
```

```python
def plot_rewards(rewards):
    plt.figure(figsize=(10, 5))
    plt.plot(rewards)
    plt.title('Total Rewards per Episode')
    plt.xlabel('Episode')
    plt.ylabel('Total Reward')
    plt.grid(True)
    plt.show()


plot_rewards(rewards_all_episodes)
```
